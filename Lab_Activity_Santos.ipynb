```
import numpy as np
import pandas as pd

ds_salaries=pd.read_csv("ds_salaries.csv")
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

salaries = ds_salaries["salary_in_usd"]
average = np.average(salaries)
std = np.std(salaries)
print("1. \nThe average salary is: $" + str(average))
print("The standard deviation is: " + str(std))

job_title = ds_salaries["job_title"]
unique = np.unique(job_title)
print("\n2. The unique job titles are:")
counter = 0
for index, element in enumerate(unique):
    if index < len(unique) - 1:
        print(element, end=', ')
    else:
        print(element, end='.')
    counter += 1
    if counter == 5:
        print()
        counter = 0

print("\n\n3. Unique Job titles with their average salary in USD and Standard Deviation")
grouped = ds_salaries.groupby('job_title')['salary_in_usd'].agg([np.mean, np.std])

# write the results to a new csv file
print(grouped)
```
